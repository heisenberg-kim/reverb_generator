```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.signal import fftconvolve
import IPython
import pyroomacoustics as pra
import soundfile as sf
import os
```

```python
# random reverberation generator

t60_list = []
room_vol_list = []

clean_path = "path to clean"
output_path = "path to reverb data"
os.makedirs(output_path, exist_ok=True)

file_list = os.listdir(clean_path)

for i in range(len(file_list)):
    
    #fs, signal = wavfile.read(os.path.join(clean_path, file_list[i]))
    signal, fs = sf.read(os.path.join(clean_path, file_list[i]))
    
    x_min = 2; x_max = 10; y_min = 2; y_max = 10; z_min = 3; z_max = 6
    x_empty_space = 0.5; y_empty_space = 0.5
    z_floor_empty_space = 0.5; z_man = 2.0
    
    x_len = np.clip(x_max*np.random.rand(1), x_min, x_max).astype('float')
    y_len = np.clip(y_max*np.random.rand(1), y_min, y_max).astype('float')
    z_len = np.clip(z_max*np.random.rand(1), z_min, z_max).astype('float')
    
    room_vol = x_len*y_len*z_len
    room_vol_list.append(room_vol)

    z_ceil_empty_space = z_len - z_man
    mic_x_loc = np.clip(x_max*np.random.rand(1), 0.0 + x_empty_space, x_len - x_empty_space).astype('float')
    mic_y_loc = np.clip(y_max*np.random.rand(1), 0.0 + y_empty_space, y_len - y_empty_space).astype('float')
    mic_z_loc = np.clip(z_man*np.random.rand(1), 0.0 + z_floor_empty_space, z_len - z_ceil_empty_space).astype('float')

    #mic_loc = np.squeeze(np.array([[mic_x_loc, mic_x_loc + 0.1], [mic_y_loc, mic_y_loc], [mic_z_loc, mic_z_loc]]), axis=2)
    mic_loc = np.squeeze(np.array([[mic_x_loc], [mic_y_loc], [mic_z_loc]]), axis=2)
    corners = np.array([[0,0], [0,y_len], [x_len,y_len], [x_len,0]]).T  # [x,y] counterclock-wise
    room = pra.Room.from_corners(corners, fs=fs, max_order=3, materials=pra.Material(0.2, 0.15), ray_tracing=True, air_absorption=True)
    room.extrude(z_len, materials=pra.Material(0.2, 0.15))
    room.set_ray_tracing(receiver_radius=0.5, n_rays=10000, energy_thres=1e-5)
    
    room.add_source([x_len/2, y_len/2, z_len/2], signal=signal)
    
    room.add_microphone(mic_loc)

    room.image_source_model()
    room.compute_rir()
    t60 = pra.experimental.measure_rt60(room.rir[0][0], fs=room.fs, plot=False)
    #print("mic position: {}, {}, {}".format(mic_x_loc, mic_y_loc, mic_z_loc))
    #print("room volume : {} m2".format(x_len*y_len*z_len))
    t60_list.append(1000*t60)

    room.simulate()
    
    #convert int 16 audio data to float32
    reverbed = room.mic_array.signals[0,:]/32767
    
    #make output data to have same length with 
    reverbed_cut = reverbed[:len(signal)]

    sf.write(os.path.join(output_path, file_list[i]), reverbed_cut, fs)
    
    if i%100 == 0:
        print("{}/{}".format(i+1, len(file_list)))
        print(f"The RT60 is {t60 * 1000:.0f} ms")
        
plt.subplot(1,2,1)
plt.hist(t60_list)
plt.xlabel("RT60(ms)")
plt.ylabel("entries")
plt.title("RT60 histogram")
plt.subplot(1,2,2)
plt.hist(np.reshape(room_vol_list, (-1)), bins=int(np.sqrt(len(file_list))))
plt.xlabel("room_volume(m^3)")
plt.ylabel("entries")
plt.title("room_volumn histogram")
#fig, ax = room.plot(img_order=3)
#fig.set_size_inches(18.5, 10.5)
```
